```python
#1 Define a list called products that contains the following items: "t-shirt", "mug", "hat", "book", "keychain".

products = ["t-shirt", "mug", "hat", "book", "keychain"]
```

```python
#2 Create an empty dictionary called inventory
inventory = {}

#3 Ask the user to input the quantity of each product available in the inventory. Use the product names from the products list as keys in the inventory dictionary and assign the respective quantities as values.
for product in products:
    quantity = int(input(f"Please enter the quantity of {product}"))
    inventory[product] = quantity
```

```python
# advanced solution for this step, simpler one follows at the end
# he system stops asking when the amount of products ordered match the amount of product available in the products list

# Create an empty set called customer_orders
customer_order = set()
max_products = len(products)
count = 0

# Ask the user to input the name of three products that a customer wants to order (from those in the products list, meaning three products out of "t-shirt", "mug", "hat", "book" or "keychain". Add each product name to the customer_orders set.
while count < max_products: # adding this because it does not make sense to keep asking the customer to add more than the available products in the inventory, hence after 5 input, the loop should terminate
    product_name = input("Please enter the name of a product you wish to order: ")
    customer_order.add(product_name)
    count += 1
    decision = input("Do you want to add another product to your order? yes or no: ")
    if decision != "yes":
        count += max_products

print(customer_order)
    
```

```python
# just a quick check of my inventory and the order of the customer
print(inventory)
print(customer_order)
```

```python
#3 Instead of updating the inventory by subtracting 1 from the quantity of each product, only do it for the products that were ordered (those in "customer_orders")

for product in inventory:
    if product in customer_order:
        inventory[product] -= 1

print(inventory)
```
